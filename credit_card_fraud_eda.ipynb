```python
#Import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#LOAD DATA:
df = pd.read_csv("D:\Santtosh Docs\Python in 100 Days\Python Data\creditcard.csv")
df.head()
```

```python
#Explore Class Balance
#df['Class'].value_counts().plot(kind='bar', title='Fraud vs Non-Fraud')
sns.countplot(x="Class", data=df)
plt.title("Fraud vs Non-Fraud Transaction Count")
plt.xlabel("Class (0 = Non-Fraud, 1 = Fraud)")
plt.ylabel("Count")
plt.show()
```

```python
#Class Balance
print("\nClass Distribution:")
print(df["Class"].value_counts())
```

```python
#Amount Distribution
sns.histplot(df['Amount'], bins=50)
```

```python
#Statistical Summary
df.describe().T
```

```python
#Transaction Amounts by Class
sns.histplot(data=df[df["Class"] == 0], x="Amount", bins=50, color="blue", label="Non-Fraud", kde=True)
sns.histplot(data=df[df["Class"] == 1], x="Amount", bins=50, color="red", label="Fraud", kde=True)
plt.title("Transaction Amount Distribution")
plt.legend()
plt.show()

```

```python
# Feature: Time (converted to hours)
df["Hour"] = (df["Time"] / 3600).round()
sns.histplot(x="Hour", data=df, hue="Class", bins=24, multiple="stack")
plt.title("Transaction Frequency by Hour")
plt.xlabel("Hour (approximate)")
plt.ylabel("Count")
plt.show()
```

```python
#Correlation Analysis
corr_matrix = df.corr()
sns.heatmap(corr_matrix, cmap="coolwarm", linewidths=0.5)
plt.title("Feature Correlation Matrix")
plt.show()
```

```python
#Focus on Most Correlated Features with Class
corr_target = corr_matrix["Class"].drop("Class").sort_values(key=abs, ascending=False)
print("\nTop Features Correlated with Fraud Class:")
print(corr_target.head(10))
```

```python
#Scatter plot for top feature vs Amount
top_feature = corr_target.index[0]
sns.scatterplot(data=df, x=top_feature, y="Amount", hue="Class", alpha=0.6)
plt.title(f"{top_feature} vs Amount Colored by Class")
plt.show()
```
